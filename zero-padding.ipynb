```python
####
import pandas as pd
import torch
data=pd.read_csv("data/temporal/X5_all_arti.csv")
tp=data.iloc[:,3:55]
```

```python
tp
```

```python
import numpy as np
tp_zero= torch.zeros(tp.shape[0], 72) 
num_rows = tp.shape[0]
num_cols = 9 * 8  # Number of elements in the "blank" matrix

data4 = np.zeros((num_rows, num_cols))
```

```python
for i in range(tp.shape[0]):
    blank = np.zeros((8, 9))  # Initialize blank matrix with zeros
    
    # Assign the value of the PC loading to the visual field
    blank[0, 3:7] = tp.iloc[i, 0:4]
    blank[1, 2:8] = tp.iloc[i, 4:10]
    blank[2, 1:9] = tp.iloc[i, 10:18]
    blank[3, 0:7] = tp.iloc[i, 18:25]
    blank[3, 8] = tp.iloc[i, 25]
    blank[4, 0:7] =tp.iloc[i, 26:33]
    blank[4, 8] = tp.iloc[i, 33]
    blank[5, 1:9] = tp.iloc[i, 34:42]
    blank[6, 2:8] = tp.iloc[i, 42:48]
    blank[7, 3:7] = tp.iloc[i, 48:52]
    
    # Blind spot
    blank[3, 7] = 0
    blank[4, 7] = 0
    
    # Flatten the matrix and assign it to data4
    data4[i, :] = blank.flatten()
```

```python
df = pd.DataFrame(data4)

# Specify the filename
filename = "data/temporal/data_72_arti.csv"

# Save the DataFrame as a CSV file
df.to_csv(filename, index=False, header=False)
```
